# **Introduction**

The sw4096.csv dataset provides data on Solar Wind Speed, with a single column of numerical values ​​that reflect variations in solar wind speed over time. Solar Wind Speed ​​is a crucial metric for understanding the dynamics of the solar wind, which is a stream of charged particles emitted by the Sun. These particles, composed mainly of electrons and protons, can affect space weather and interact with the Earth's magnetic field. Earth, causing phenomena such as auroras and geomagnetic storms.

# **Objective**

The objective of this work is to use the sw4096.csv dataset to predict future solar wind speeds using machine learning techniques. Specifically, the focus is to employ a Long Short-Term Memory (LSTM) neural network to predict solar wind speeds for the next five time periods. This forecast aims to provide insights into future variations in the solar wind, helping to anticipate changes in space weather.

# **Methodology**

To achieve the proposed objective, the methodology includes the following steps:

1. Data Preprocessing: Cleaning and preparing data from the sw4096.csv dataset, including normalizing values ​​and dividing data into training and testing sets.

2. LSTM Model Construction: Development of an LSTM model to capture temporal dependencies in solar wind speed data.

3. Topology Exploration: The experiment uses five different LSTM topologies to identify the configuration that best fits the data. These topologies vary in terms of layers and units within each layer, providing a comprehensive approach to finding the most effective architecture.

4. Training and Validation: Each of the five topologies is trained and validated using appropriate techniques to optimize performance and avoid overfitting. The selection of the best parameters and hyperparameters is performed for each topology.

5. Forecasting: Using the most effective LSTM topology to predict solar wind speeds for the next five time periods based on available historical data.

# **Experiments and Results**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout, Input
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

drive.mount('/content/drive')
path = '/content/drive/MyDrive/Att_mlp/sw4096.csv'
df = pd.read_csv(path, header=None)
df.fillna(0, inplace=True)

data = df.values

def create_sequences(data, sequence_length=10, forecast_steps=5):
    X, y = [], []
    for i in range(len(data) - sequence_length - forecast_steps + 1):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length:i + sequence_length + forecast_steps])
    return np.array(X), np.array(y)

sequence_length = 10
forecast_steps = 5

X, y = create_sequences(data, sequence_length, forecast_steps)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=30)

scaler_X = MinMaxScaler()
X_train = scaler_X.fit_transform(X_train.reshape(-1, 1)).reshape(X_train.shape)
X_test = scaler_X.transform(X_test.reshape(-1, 1)).reshape(X_test.shape)

scaler_y = MinMaxScaler()
y_train = scaler_y.fit_transform(y_train.reshape(-1, forecast_steps)).reshape(y_train.shape)
y_test = scaler_y.transform(y_test.reshape(-1, forecast_steps)).reshape(y_test.shape)

def create_model(sequence_length, forecast_steps, units, dropout_rate=None):
    model = Sequential()
    model.add(Input(shape=(sequence_length, 1)))
    model.add(LSTM(units, return_sequences=True))
    model.add(LSTM(units, return_sequences=False))
    if dropout_rate:
        model.add(Dropout(dropout_rate))
    model.add(Dense(forecast_steps))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def train_and_evaluate_rnn(model, epochs, batch_size):
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2, verbose=1)
    loss = model.evaluate(X_test, y_test)
    print(f'Mean Squared Error: {loss:.5f}')
    y_pred = model.predict(X_test)
    y_test_squeezed = np.squeeze(y_test)
    y_test_flat = y_test_squeezed.flatten()
    y_pred_flat = y_pred.flatten()
    mse = mean_squared_error(y_test_flat, y_pred_flat)
    mae = mean_absolute_error(y_test_flat, y_pred_flat)
    r2 = r2_score(y_test_flat, y_pred_flat)
    print(f'\nMean Squared Error: {mse:.5f}')
    print(f'Mean Absolute Error: {mae:.5f}')
    print(f'R^2 Score: {r2:.2f}')
    print('-' * 50)

    return history, y_test_squeezed, y_pred

configurations = [
    {'units': 64, 'dropout': None, 'epochs': 10, 'batch_size': 32},
    {'units': 128, 'dropout': 0.2, 'epochs': 20, 'batch_size': 32},
    {'units': 64, 'dropout': 0.3, 'epochs': 30, 'batch_size': 64},
    {'units': 128, 'dropout': None, 'epochs': 40, 'batch_size': 16},
    {'units': 128, 'dropout': 0.4, 'epochs': 50, 'batch_size': 32},
]

for i, config in enumerate(configurations):
    print(f"Configuração {i+1}")
    model = create_model(sequence_length, forecast_steps, config['units'], config['dropout'])
    history, y_test_squeezed, y_pred = train_and_evaluate_rnn(model, config['epochs'], config['batch_size'])

    fig, axs = plt.subplots(2, 2, figsize=(18, 12))

    axs[0, 0].plot(history.history['loss'], label='Loss (Treinamento)')
    axs[0, 0].plot(history.history['val_loss'], label='Loss (Validação)')
    axs[0, 0].set_title('Perda Durante o Treinamento')
    axs[0, 0].set_xlabel('Época')
    axs[0, 0].set_ylabel('Perda')
    axs[0, 0].legend()

    errors = y_test_squeezed - y_pred
    axs[0, 1].scatter(range(len(errors.flatten())), errors.flatten(), alpha=0.6)
    axs[0, 1].set_title('Dispersão dos Erros de Previsão')
    axs[0, 1].set_xlabel('Índice')
    axs[0, 1].set_ylabel('Erro')

    axs[1, 0].bar(np.arange(len(errors.flatten())), errors.flatten())
    axs[1, 0].set_title('Análise dos Erros de Previsão')
    axs[1, 0].set_xlabel('Índice')
    axs[1, 0].set_ylabel('Erro')

    axs[1, 1].plot(np.arange(len(y_train.flatten())), y_train.flatten(), label='Treinamento (Real)', color='blue')
    axs[1, 1].plot(np.arange(len(y_train.flatten()), len(y_train.flatten()) + len(y_test_squeezed.flatten())), y_test_squeezed.flatten(), label='Teste (Real)', color='orange')
    axs[1, 1].plot(np.arange(len(y_train.flatten()), len(y_train.flatten()) + len(y_pred.flatten())), y_pred.flatten(), label='Previsto', color='red')
    axs[1, 1].set_title('Previsões Finais')
    axs[1, 1].set_xlabel('Índice')
    axs[1, 1].set_ylabel('Amplitude')
    axs[1, 1].legend()

    fig_name = f'model_{config["units"]}_units_dropout_{config["dropout"]}_epochs_{config["epochs"]}.png'
    plt.savefig(fig_name)
    plt.show()
```
